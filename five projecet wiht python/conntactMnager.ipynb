```python
import json
import os

class ContactManager:
    def __init__(self, path="-"):
        self.contact_list = []
        self.path = path

        if path != "-" and os.path.exists(path):
            print("Loading previous contacts...")

            with open(path, "r") as f:
                try:
                    self.contact_list = json.load(f)
                    print("LOADED...")
                except json.JSONDecodeError:
                    print("File empty or corrupted. Starting new list.")
                    self.contact_list = []

    def add(self, name, number):
        self.contact_list.append({
            "name": name,
            "number": number
        })
        print("Contact added.")

    def search(self, name):
        result = []
        for item in self.contact_list:
            if name.lower() in item["name"].lower():
                result.append(item)

        print("Search result:", result)

    def backup(self):
        with open("./contact_list.json", "w") as f:
            json.dump(self.contact_list, f, indent=4)
        print("Backup saved successfully.")

    def show_all(self):
        print("Your contacts:")
        for contact in self.contact_list:
            print(contact)
```

```python
my_contacts = ContactManager(path="./contact_list.json")

my_contacts.show_all()
```

```python
my_contacts = ContactManager(path="./contact_list.json")

my_contacts.add("Ali", "0788888888")
my_contacts.backup()
my_contacts.show_all()
```
